# Annif Metrieken
Hier staan alle metrieken vastgelegd van gebruikte Annif backends voor Demosaurus.

## Samenvatting

Voor onze toepassing van Annif is het belangrijk dat we de Recall waarde proberen te maximaliseren.

- Brinkman zaaktrefwoorden backend recall: **0,54**
- Brinkman vormtrefwoorden backend recall: **0,94**
- CBK genretrefwoorden backend recall:     **0,79**

Bovenstaande recall waardes zijn berekend bij een limiet van 20 trefwoorden.
Dat betekend dat bv. voor de Brinkman zaaktrefwoorden backend dat van de 20 zaaktrefwoorden die door Annif worden voorgesteld er 54% relevant zijn.

## Brinkman zaaktrefwoorden backend

De Brinkman zaaktrefwoorden backend is een _Ensemble_ (combinatie) bestaande uit _Omikuji_ en _FastText_ backends met een ratio van 3:1; de output van omikuji telt 3x zo zwaar mee in de uiteindelijke output.
De ensemble backend heeft een _recall_ waardee van 0,54 bij een limiet van 20, wat betekend dat van de 20 zaaktrefwoorden die Annif voorstelt er 54% relevant zijn.
